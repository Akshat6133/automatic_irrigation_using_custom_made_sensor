```python
import matplotlib.pyplot as plt

# Data for sensor a1
water_content = ['20ml', '40mL', '60mL', '80mL', '100mL', '120mL']
pF_20kHz = [17.6, 17.17, 18.1, 20.2, 21.5, 19.8]

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(water_content, pF_20kHz, marker='o', color='b', linestyle='-')

# Adding labels and title
plt.xlabel('Water Content')
plt.ylabel('20kHz(pF)')
plt.title('Sensor a1 Data')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Displaying the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
