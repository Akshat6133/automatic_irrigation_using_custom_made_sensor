# A1

```python
import matplotlib.pyplot as plt

# Data for sensor a1
water_content = ['20ml', '40mL', '60mL', '80mL', '100mL', '120mL']
pF_2kHz = [18.26, 17.62, 19.43, 18.9, 19.7, 19.8]
pF_20kHz = [17.6, 17.17, 18.1, 20.2, 21.5, 19.8]
pF_200kHz = [17.11, 16.71, 17.4, 19.4, 20.3, 20.4]
pF_1M = [16.66, 16.28, 16.9, 18.3, 19, 22.4]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(water_content, pF_2kHz, marker='o', label='2kHz(pF)')
plt.plot(water_content, pF_20kHz, marker='o', label='20kHz(pF)')
plt.plot(water_content, pF_200kHz, marker='o', label='200kHz(pF)')
plt.plot(water_content, pF_1M, marker='o', label='1M(pF)')

# Adding labels and title
plt.xlabel('Water Content')
plt.ylabel('pF Value')
plt.title('Sensor a1 Data')
plt.legend()

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Displaying the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
